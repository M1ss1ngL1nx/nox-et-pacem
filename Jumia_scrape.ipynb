# Web Scraping on Jumia.


```python
#This is the link to the jumiaa site
#'https://www.jumia.com.gh/men-sneakers/'
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.jumia.com.gh/men-sneakers/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
}

# Send a GET request to the URL
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(response)
```

```python
print(soup)
```

```python
product = soup.find_all('a', class_='core')
```

```python
print(product)
```

```python
import pandas as pd
```

```python
data  = []

for items in product:
    product_name = items.find('h3', class_='name').text.strip()
    image_url = items.find('img')['data-src']
    product_price = items.find('div', class_='prc').text.strip()
    product_discount = items.find('div', class_='bdg')
    product_rating = items.find('div', class_='stars _s')
    product_old_price = items.find('div', class_='old').text.strip()
    delivery_method = items.find('span', class_='bdg _xs camp')

    #product_shipping = items.find('svg')['aria-label']
    
    if delivery_method:
        delivery_method = delivery_method.text.strip()

    if product_discount:
        discount = product_discount.text.strip()
        
    else:
        discount = 'No discount'
    if product_rating:
        product_rating = product_rating.text.strip()
        
    #split Product into Product name and color
    product_parts = product_name.split('-')

    product_name = product_parts[0].strip()
    color = product_parts[-1].strip() if len(product_parts) > 1 else None
    
    
    
    print('Product Name:', product_name)
    print('Color:', color)
    print('Image_URL:', image_url)
    print('Price:', product_price)
    print('Old_Price:', product_old_price)
    print('Discount:', discount)
    print('Rating:', product_rating)  
    
    print('Delivery_Method:', delivery_method)
    

    
    #print('Shipping:', product_shipping)

    # Create a DataFrame from the extracted information
    product_data = {
    'Product': product_name,
    'Color':color,
    'Price': product_price,
    'Old_Price': product_old_price,
    'Discount': discount,
    'Rating': product_rating,    
    'Delivery_Method': delivery_method,
    'Image_URL': image_url
    }

    # Append the product data to the empty list
    data.append(product_data)

    # Create a DataFrame from the list of dictionaries
    df = pd.DataFrame(data)
```

```python
df
```

```python
# Save the DataFrame to a CSV file
df.to_csv('jumia_data.csv', index=False)
```
